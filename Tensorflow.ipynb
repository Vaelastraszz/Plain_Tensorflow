# Création d'un réseau de neuronnes en Plain Tensorflow

```
n_inputs = X.shape[1]
n_hidden1=200
n_hidden2=100
n_output=6
```

```
from datetime import datetime 

now = datetime.utcnow().strftime("%Y%m%d%H%M%S")
root_logdir = "/content/gdrive/My Drive/tf_logs"
#root_logdir ='./log'
logdir = "{}/run-{}/".format(root_logdir,now)
```

<h2> Phase de construction du reseau de neuronnes </h2>

```
X_tf = tf.placeholder(tf.float32, shape=(None, n_inputs), name="X_tf")
y_tf = tf.placeholder(tf.int64, shape=(None), name="y_tf")
```

Utilisations de noeuds placeholder pour stocker notre training set et notre target training set, 2D tensor pour X et 1D tensor pour y avec None pour indiquer qu'on ne sait pas encore la taille du batch en entrée de X et y 

```
def neuron_layer(X, n_neurons, name, activation=None):
    with tf.name_scope(name):
        n_inputs = int(X.get_shape()[1])
        stddev = 2/np.sqrt(n_inputs + n_neurons)
        init = tf.truncated_normal((n_inputs, n_neurons),stddev=stddev)
        W = tf.Variable(init, name="kernel")
        b = tf.Variable(tf.zeros([n_neurons]), name= "bias")
        Z = tf.matmul(X,W) + b
        tf.summary.histogram("weights", W)
        if activation is not None:
            tf.summary.histogram("activation", activation(Z))
            return activation(Z)
        else:
            return Z
                        
```

Création d'une fonction pour créer nos layer de neuronnes : 2 hidden layers + un layer d activation softmax pour la classification :

<ul>
  <li> On définit nos input avec la shape[1] de la matrice d'observations en entrée pour récupérer le nb d'instances</li>
  <li> On initialise nos vecteurs poids de dimension 2 nb_input/ nb de neuronnes + on utilise un distribution gaussienne tronquée pour avoir des poids pas trop grands pour ne pas ralentir le modèle, les vecteurs poids seront initialisés de manière aléatoire pour éviter une symétrie de distribution des poids pour que la descente de gradient fonctionne correctement.</li>
  <li>On créé un vecteur poids 1 dimension de la taille du nb de neuronne du layer, initialisé à 0</li>
  <li>On va multiplier la somme des poids des inputs et ajouter le bias pour chaque neuronne du layer</li>
    <li> Utilisation d'une fonction d'activation pour le layer si besoin (i.e ReLU / softmax . tanh...)</li>
</ul>





```
with tf.name_scope("dnn"):
    hidden1 = neuron_layer(X_tf, n_hidden1, name="hidden1", activation=tf.nn.relu)
    hidden2 = neuron_layer(hidden1, n_hidden2, name="hidden2", activation=tf.nn.relu)
    logits = neuron_layer(hidden2, n_output, name="outputs")
```

Création de nos layer grace a la fonction de création de layer qu'on a fait au dessus, logits est l'output du réseau de neuronne avant de passer dans la fonction softmax pour raison d'optimisation. On pourra plus tard utilisé la fonction tf.layers.dense pour que tensorflow s'occupe de créer les layers à notre place mais ici on va chercher à réaliser le réseaux de neuronne au plus bas niveau de création possible de tensorflow


*   exemple :  tf.layers.dense(X, n_hidden1, name="hidden1", activation=tf.nn.relu)





```
with tf.name_scope("loss"):
    xentropy = tf.nn.sparse_softmax_cross_entropy_with_logits(labels=y_tf, logits=logits)
    loss = tf.reduce_mean(xentropy, name="loss")
```

On va définir ici la fonction de perte du réseau de neuronne pour la multiclassification qui sera la cross-entropy qui va pénaliser les instances qui n'ont pas été atrribuées la classe voulue. La cross-entropy va donc nous renvoyer un vecteur l'entropie de chaque instance pour savoir à quel point elle a été attribué a la bonne instance et la mean entropy va donc calculer l'erreur moyenne de l'entropie de toutes les instances du batch 

```
learning_rate = 0.01

with tf.name_scope("train"):
    optimizer = tf.train.GradientDescentOptimizer(learning_rate)
    training_op = optimizer.minimize(loss)
```

On va ici s'occuper de la descente de gradient qui permettra de réduire la fonction de perte qu'on a déclaré plus haut

```
with tf.name_scope("eval"):
    correct = tf.nn.in_top_k(logits, y_tf, 1)
    accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))
```

Ici on créé lla partie pour évaluer la précision du modèle en observant si le logit le plus haut correspond à la classe visée et on effectue ensuite la moyenne de précision sur l'ensemble des instances du batch 

```
init = tf.global_variables_initializer()
```

Création du noeud d'initialisation des variables du graphe

```
entrop_resume = tf.summary.scalar("cross_entropy", loss)
accuracy_resume = tf.summary.scalar("accuracy", accuracy)

merged = tf.summary.merge([entrop_resume, accuracy_resume])
```

Création d'opération tensorboard pour observer 

<h2> Execution du réseau de neuronnes </h2>

```
n_epochs = 60
batch_size = 10
```

```
def next_batch(num, data, labels):

    '''
    Return a total of `num` random samples and labels. 
    '''
    idx = np.arange(0 , len(data))
    np.random.shuffle(idx)
    idx = idx[:num]
    data_shuffle = data.iloc[idx]
    labels_shuffle = labels.iloc[idx]
    #labels_shuffle = np.asarray(labels_shuffle.values.reshape(len(labels_shuffle), 1))

    return data_shuffle, labels_shuffle
```
